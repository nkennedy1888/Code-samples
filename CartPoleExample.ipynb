```python
import gym

import numpy as np

import random

env=gym.make('CartPole-v1')




done=False




alpha = 0.1

gamma = 0.6

epsilon = 0.1

s=env.reset()

counter_of_consecutive_success = 0

sum_of_rewards = 0

w, h = 200, 100

q_table = [[0.0 for x in range(h)] for y in range(w)]




for i in range(1,1001):

    state=env.reset()

    env.render()

    done=False




    pr=0

    action=0

    while not done:




        if random.uniform(0, 1) < epsilon:

            action = env.action_space.sample()

            pr=action

        else:

            action = np.argmax(q_table[int(state[0])])

            action=pr




next_state, reward, done, info = env.step(action)

print("Current reward: ", reward)

env.close()





old_value = q_table[int(state[0])][int(action)]







next_max = q_table[int(next_state[0])]

new_value = (1 - alpha) * old_value + alpha * (reward + gamma * next_max[0])

q_table[int(state[0])][int(action)] = new_value

state = next_state

sum_of_rewards+=reward

print("At ",i,"-th episode the average reward is ",np.mean(sum_of_rewards))

if np.mean(sum_of_rewards) >= 195.0:

    print(i)

    print("Current Mean reward: ",np.mean(sum_of_rewards))

    counter_of_consecutive_success+=1

    print("Current counter of consecutive success ",counter_of_consecutive_success)

if np.mean(sum_of_rewards) >= 195 and counter_of_consecutive_success >= 100:

    print("Problem solved at ", i, "-th episode with average reward ", np.mean(sum_of_rewards))

    break

if np.mean(sum_of_rewards) >= 195 and counter_of_consecutive_success >= 100:

    break
```
